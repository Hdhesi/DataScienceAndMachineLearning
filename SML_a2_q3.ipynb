_3a) importing necessary libraries and modules:_

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
random.seed(0)
```

_3b) creating an array of coefficients based on my student number, and defining a polynomial._

```python
# student number = 0371098
# a = 8, b = 9, c = 0, d = 1

coefs = [1,0,9,8]

def polynomial(coefs,x):
    n = len(coefs)
    for i in range (1,n):
        p = sum([(coefs[i]+1)*x**(i)])
    polynomial = p + coefs[0]
    return polynomial

def draw_polynomial(coefs):
    n = len(coefs)
    x = np.linspace(-5, 5, 1000)
    plt.ylim(-20,20)
    plt.plot((x, sum([(coefs[i]+1)*x**(i) for i in range (1,n)]) + coefs[0] ), linestyle='-', color='black')

#draw_polynomial(coefs)

#x = 1
#p = polynomial(coefs,x)
#print(p)
```

```python
X = []
Y = []
for i in range(40):
    x = random.uniform(-5,5)
    y = polynomial(coefs, x) + random.gauss(0,2)
    X.append(x)
    Y.append(y)
```

```python
plt.scatter(X, Y)
```

## Training a polynomial regression model

```python
import turicreate as tc
```

```python
data = tc.SFrame({'x':X, 'y':Y})
data
```

### Adding columns to our dataset corresponding to $x^2, x^3, \ldots, x^{200}$

```python
for i in range(2,200):
    string = 'x^'+str(i)
    data[string] = data['x'].apply(lambda x:x**i)
data
```

## Splitting data into training and testing

_3c)_ _**75% of data is used for training**_ _(rest 25% goes for testing)._

```python
train, test = data.random_split(.75, seed=0)
```

```python
def display_results(model):
    coefs = model.coefficients
    print("Training error (rmse):", model.evaluate(train)['rmse'])
    print("Testing error (rmse):", model.evaluate(test)['rmse'])
    plt.scatter(train['x'], train['y'], marker='o')
    plt.scatter(test['x'], test['y'], marker='^')
    draw_polynomial(coefs['value'])
    plt.show()
    print("Polynomial coefficients")
    print(coefs['name', 'value'])
```

## Training a polynomial regresssion model with both L1 and L2 regularization

_3d) creating a model called_ _**model_elastic_reg**_ _that uses an L1 penalty of 0.1 and an L2 penalty of 0.1_.

```python
model_elastic_reg = tc.linear_regression.create(
    train, target='y', l1_penalty=0.1, l2_penalty=0.1, verbose=False, validation_set=None)
```

_3d) displaying the result of the above model_.

```python
display_results(model_elastic_reg)
```

_3e) comparison of the training and testing error of model_elastic_reg:_   
_**the training error as well as the testing error both are extremely high**_, _it is a case of_ _**underfitting**_.

## Predictions for each point in the testing data

_3f) displaying the predictions made by model_elastic_reg_

```python
predictions = test['x', 'y']
predictions['elastic reg'] = model_elastic_reg.predict(test)
```

```python
predictions
```

## Training another polynomial regresssion model 

_3g) creating a new model named_ _**optimal_model**_, _by experimenting with differet values of L1 and L2 penalties and coming up with one which gives better results than model_elastic_reg._ _**I have used an l1 penalty of 0.0 and an l2 penalty of 0.0001**._ _The model gave very less training and testing error this time._

```python
optimal_model = tc.linear_regression.create(
    train, target='y', l1_penalty=0.0, l2_penalty=0.0001, verbose=False, validation_set=None)
```
