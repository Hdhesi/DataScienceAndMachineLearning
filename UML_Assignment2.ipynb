```python
# importing necessary libraries

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans 
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
```

```python
# reading the wine dataset in a pandas dataframe df_wine

df_wine = pd.read_csv("wine_data.csv")
```

```python
# printing small part of the df_wine dataframe 

df_wine.head()
```

```python
# visualising the features by plotting the OD Reading feature against the Proline feature

plt.scatter(df_wine["OD_read"], df_wine["Proline"])
plt.xlabel("OD_Reading")
plt.ylabel("Proline")
plt.title("Wine Dataset")

plt.show()
```

### Observation

From the above scatter plot, **it can be observed that the clusters of the data are very hard to locate bu just looking at the plot**. And, the relationship between the two features **seems positive**, however, it is **not very strong**.

### Main steps of - K-means Clustering

1. First Pick "k" centroids, where "k" is the expected number of clusters. 
2. Then, place the "k" centroids at random places among the existing data points.
3. Next, calculate the distance from each centroid to each point in the data.
4. Then based on the distance, all the points are grouped with their nearest clusters.
5. Calculate the mean/average data point in the group and move the previous centroid to the mean location.
6. Repeat the process until convergence (when the value of centroid stops changing) or until maximum iteration limit has been achieved.

### Main steps of - Hierarchical Clustering

1. Given n sample data points, view each point as an individual "cluster" with just that one point as a member (the centroid).
2. Calculate the distance between all the possible pairs of the centroids of all the clusters in the data. 
3. Group the closest clusters/points together in pairs.
4. Repeat Step 2 and Step 3 until you get a single cluster containing all the data points in the set.
5. Plot a dendrogram to show how the data has come together in a hierarchical structure. 
6. Decide what level/depth you want to create the clusters at.


### Similarity between K-means Clustering and Hierarchical Clustering

Both models return the clusters in the dataset based on the similarity between the points using distance between them.

### Differences between K-means Clustering and Hierarchical Clustering

1. In K- means, the value of k (number of centroids) is chosen by us. However, in Hierarchical, each point is considered as a cluster itself with that point as the centroid and the depth can be specified to find the required clusters.
2. In K-means, the minimum distance between the point and the centroid is used to cluster points, however, in Hierarchical, minimum distance between clusters, maximum distance between clusters, average distance between clusters, or distance between two centroids can also be considered.
3. The K-means Clustering process ends when it reaches convergence (when the value of centroid stops changing), or when the maximum set iterations are achieved. However, the Hierarchical Clustering process ends when a single cluster containing all the datapoints is obtained.

## K-means Clustering on the Wine data with 3 Clusters

```python
# sklearn implementation of k-means clustering

kmc = KMeans(3)
kmc_clusters = kmc.fit_predict(df_wine)
```

```python
# Plotting the predicted clusters of k_means

plt.scatter(df_wine["OD_read"], df_wine["Proline"], c = kmc_clusters, cmap='tab20b')
plt.xlabel("OD_read")
plt.ylabel("Proline")
plt.title("Wine Clusters for K-means Clustering")

plt.show()
```

```python
# Silhoutte Score for k-means clustering

print("Silhoutte_score for K-means Clustering on the Wine Data: ", silhouette_score(df_wine, kmc_clusters))
```

## Hierarchical (Agglomerative) Clustering on the Wine data with 3 Clusters

```python
# sklearn implementation of hierarchical clustering

ahc = AgglomerativeClustering(3, linkage='average')
ahc_clusters = ahc.fit_predict(df_wine)
```

```python
# Plotting the predicted clusters of hierarchical slustering

plt.scatter(df_wine["OD_read"], df_wine["Proline"], c = ahc_clusters)
plt.title("Wine Clusters from Heirarchical Clustering")
plt.xlabel("OD_read")
plt.ylabel("Proline")

plt.show()
```

```python
# Silhoutte Score for hierarchical clustering

print("Silhoutte_score for Hierarchical Clustering on the Wine Data: ", silhouette_score(df_wine, ahc_clusters))
```
