### Question 2

a)

```python
# creating a column in the dataframe as cumulative to find if a student passed or failed

df["cumulative"] = df["G1"] + df ["G2"] + df ["G3"]

# creating a function that uses the above cumulative score to find the result (0 - pass, 1 - fail)

def result(cumulative):
    if cumulative >= 35:
        result = 1
    else:
        result = 0 
    return result   
```

```python
# creating a column 'pass' in the dataframe and using apply function to add the result of the above function to store 0 and 1 values in the column

df["pass"] = df["cumulative"].apply(result)
```

b)

```python
# dropping the Gi columns and also the cumulative column that I created above 

df = df.drop(["G1", "G2", "G3", "cumulative"], axis = 1)
```

## **Part 2 - Fitting a Decision Tree**

### Question 5

a)

```python
# importing tree from sklearn

from sklearn import tree
```

b)

```python
# creating a decision tree

dt = tree.DecisionTreeClassifier(criterion = 'entropy', max_depth = 5)
```

c)

```python
# traning the decision tree with the training data

dt.fit(train_att, train_pass )
```

### Question 6

a)

```python
# importing graphviz

import graphviz
```

b)

```python
# creating a visual representation of the decision tree created above

dot_data = tree.export_graphviz(dt, out_file = None, label = "all", impurity = False, proportion = True,\
                               feature_names = list(train_att), class_names = ["fail", "pass"], \
                               filled = True, rounded = True)
graph = graphviz.Source(dot_data)
graph
```

c)

```python
# saving the decision tree

tree.export_graphviz(dt, out_file = "student-performance.dot", label = "all", impurity = False, proportion = True,\
                     feature_names = list(train_att), class_names = ["fail", "pass"], filled = True, rounded = True)
```

### Question 7

a)

```python
# getting the score of the decsion tree

print('Testing Score: ', dt.score(test_att, test_pass))
```

b)

```python
# importing cross_val_score from sklearn

from sklearn.model_selection import cross_val_score
```

c)

```python
# cross validating the decsion tree around the testing data 

scores = cross_val_score(dt, test_att, test_pass, cv = 5)
```

d)

```python
# printing out the cross validation score obtained as score.mean, +/- two standard deviations

print("mean: ", scores.mean())
      
print("+/- two standard deviations: ", scores.std()*2)

print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std()*2))
```

### Question 9

a)

```python
# importing numpy to deal with arrays

import numpy as np

# creating an empty numpy array to store the numerical data obtained above 

depth_acc = np.empty((19,3), float)
```

b)

```python
# filling the numpy array 
i = 0

for max_depth in range(1, 20):
  dt = tree.DecisionTreeClassifier(criterion = "entropy", max_depth = max_depth)
  scores = cross_val_score(dt, test_att, test_pass, cv = 5)
  depth_acc[i, 0] = max_depth 
  depth_acc[i, 1] = scores.mean()
  depth_acc[i, 2] = scores.std()*2
  i += 1

  print("Max depth: %d, Accuracy: %0.2f (+/- %0.2f)" % (max_depth, scores.mean(), scores.std()*2))
```

The **best accuracy** value is **74 (+/- 0.19)** at maximum depth = 19 and **74 (+/- 0.11)** at maximum depth = 11.


```python
depth_acc
```

c)

```python
# importing matplotlib to generate an error plot graph for max_depth vs. mean accuracy

import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.errorbar(depth_acc[:, 0], depth_acc[:, 1], yerr = depth_acc[:, 2])

plt.show()
```
